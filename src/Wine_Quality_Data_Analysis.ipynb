# **Wine Quality Dataset: Data Analysis**

This Jupyter Notebook provides an in-depth analysis of a wine quality dataset. It focuses on understanding the relationship between various chemical properties of wines and their quality ratings. It also includes data cleaning, exploratory data analysis (EDA), and correlation studies to identify any key factors that may influence wine quality.

```python
import requests
import io
import zipfile
import pandas as pd
```

## **1. Data Loading**

Imports necessary libraries and loads the wine quality dataset.

```python
# Load and open the zipped dataset
r = requests.get("https://archive.ics.uci.edu/static/public/186/wine+quality.zip")
z = zipfile.ZipFile(io.BytesIO(r.content))
with z.open("winequality-red.csv") as f:
   # Read the dataset
   wine_quality = pd.read_csv(f,delimiter=";")
   # Display the dataset
   print(wine_quality.head())
```

## **2. Exploratory Data Analysis (EDA)**

Summarizes the main characteristics of the dataset using descriptive statistics and visualizations.

```python
wine_quality.cov()
```

```python
wine_quality.corr()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="white")

# Generate mask for the upper triangle
mask = np.triu(np.ones_like(wine_quality.corr(), dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(7, 5))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(np.abs(wine_quality.corr()), mask=mask, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
# https://seaborn.pydata.org/examples/many_pairwise_correlations.html
```

## **3. Correlation Analysis**

Analyzes the correlation between different variables and wine quality to understand which factors most influence quality.

```python
# Obtain the correlation matrix using Pandas
correlation_matrix = wine_quality.corr()

# Identify the attribute with the highest absolute correlation with quality
quality_corr = correlation_matrix["quality"].drop("quality")    # Drop self-correlation of "quality"
max_corr_attribute = quality_corr.abs().idxmax()
max_corr_value = quality_corr[max_corr_attribute]

# Determine if the association is positive or negative
association = "positive" if max_corr_value > 0 else "negative"

# Identify the two non-target attributes with the highest correlation coefficient
# Start by removing the 'quality' row and column from the correlation matrix
corr_without_quality = correlation_matrix.drop("quality", axis=1).drop("quality", axis=0)

# Locate the highest non-diagonal correlation
np.fill_diagonal(corr_without_quality.values, 0)    # Zero out the diagonal to ignore self-correlations
location = np.where(np.abs(corr_without_quality) == np.abs(corr_without_quality).max().max())
attributes = corr_without_quality.index[location[0][0]], corr_without_quality.columns[location[1][0]]
highest_corr_value = corr_without_quality.values[location[0][0], location[1][0]]

print(f"Attribute with the highest abs(correlation) with quality: {max_corr_attribute} ({max_corr_value:.2f}) - {association}")
print(f"Two non-target attributes with the highest correlation coefficient: {attributes[0]} and {attributes[1]} ({highest_corr_value:.2f})")
```
